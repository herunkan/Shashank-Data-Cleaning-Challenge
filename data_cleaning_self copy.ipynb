```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('data_cleaning_challenge.csv')
```

```python
df.head(20)
df.drop(columns=['Unnamed: 9', 'Unnamed: 10'], inplace=True)
```

```python
start_indices = df[df['Row Type'].str.contains('first name', na=False)].index.tolist()
print("Start indices for each person:", start_indices)
```

```python
for idx, start_idx in enumerate(start_indices):
    if idx < len(start_indices) - 1:
        end_idx = start_indices[idx + 1]
        df.loc[start_idx:end_idx, 'Person_index'] = idx + 1
    else:
        df.loc[start_idx:, 'Person_index'] = idx + 1

df['Person_index'] = df['Person_index'].astype(int)
```

```python
df = df[df['Row Type'].notna()]
df.head(40)
```

```python
df.drop(df[df['Row Type'].isin(['Average', 'Maximum', 'Std.Dev.', 'Total'])].index, inplace=True )
df.drop(df[df['Row Type'] == 'Row Type'].index, inplace = True)
df.head(40)
```

```python
seperated_df = df[df['Row Type'].str.contains('first name:')]
seperated_df = seperated_df[['Row Type', 'Iter Number', 'Power1', 'Person_index']]
seperated_df['Row Type'] = seperated_df['Row Type'].str.split(':').str[1]
seperated_df['Iter Number'] = seperated_df['Iter Number'].str.split(':').str[1]
seperated_df['Power1'] = seperated_df['Power1'].str.split(':').str[1]
seperated_df.rename(columns={'Row Type': 'First name', 'Iter Number' : 'Last Name', 'Power1': 'Date'}, inplace=True)
seperated_df
```

```python
df.drop(df[df['Row Type'].str.contains('first name:')].index, inplace = True)
df
```

```python
df = df.merge(seperated_df, how='outer', on = 'Person_index')
new_column_order = ['First name', 'Last Name', 'Person_index', 'Date', 'Row Type', 'Iter Number', 'Power1', 'Speed1', 'Speed2', 'Electricity',
       'Effort', 'Weight', 'Torque']
df = df[new_column_order]
```
